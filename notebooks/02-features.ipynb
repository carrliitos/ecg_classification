```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw = "../data/interim/ecg-id-database-2.0.0/Person_01_rec_10_interim.csv"
```

```python
with open(data_raw, "r") as in_file:
    df = pd.read_csv(in_file)
```

```python
print(df)
```

```python
ecg_values = df['ECG_I_filtered']
wave_labels = df['wave_label']
```

```python
plt.figure(figsize=(13, 10))

plt.scatter(range(len(ecg_values)), ecg_values, c=wave_labels.map({'p': 'red', 'q': 'blue', 'r': 'green', 's': 'orange', 't': 'purple'}))

plt.xlabel('ECG Data Point')
plt.ylabel('ECG Value')
plt.title('ECG Values with Wave Labels')

plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) for label, color in {'p': 'red', 'q': 'blue', 'r': 'green', 's': 'orange', 't': 'purple'}.items()])

plt.show()
```
